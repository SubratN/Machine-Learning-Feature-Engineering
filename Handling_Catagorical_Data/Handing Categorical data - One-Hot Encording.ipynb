### OneHotEncording with Top categories

- As we can see 'brand' column contains many categories will encord it based on top categories

```python
df['brand'].value_counts()
```

```python
df['brand'].nunique()
```

```python
counts = df['brand'].value_counts()

thershold = 100
```

```python
repl = counts[counts < thershold].index
```

```python
df_new = pd.get_dummies(df['brand'].replace(repl,'uncommon'))
```

```python
df = pd.concat([df,df_new],axis=1)
```

```python
df.drop('brand',axis=1,inplace=True)
```
